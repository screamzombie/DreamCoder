```python

import numpy as np 
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python

dataset  = pd.read_csv("processed_results.csv")
```

```python

dataset = pd.get_dummies(dataset)

```

```python

dataset.drop("Parkinson", axis =1 ).corrwith(dataset.Parkinson).plot.bar(figsize = (20, 10), title = "high income corr with")
```

```python

y = dataset["Parkinson"]
X = dataset.drop("Parkinson", axis= 1)
```

```python

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X) 
```

```python


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size  = 0.2)
X_train.shape
```

```python

from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()

model.add(Dense(output_dim=16,input_dim =12,activation='relu'))

model.add(Dense(output_dim=24,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(20,activation='relu'))
model.add(Dense(24,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(24,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(24,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(24,activation='relu'))


model.add(Dense(1,activation='sigmoid'))

model.summary() 
```

```python

model.compile(optimizer="adam", loss= "binary_crossentropy", metrics= ["accuracy"])
```

```python

history=model.fit(X_train, y_train, batch_size= 15,nb_epoch=1000)
```

```python
import matplotlib.pyplot as plt
```

```python
# history=model.fit()

plt.plot(history.history['loss'],"ro-")

# plt.plot(history.history['val_loss'])

plt.title("model loss")

plt.ylabel("loss")

plt.xlabel("epoch")

plt.legend(["train","test"],loc="upper left")

plt.show()
```

```python


plt.plot(history.history['acc'],"bo--")



plt.title("model acc")

plt.ylabel("acc")

plt.xlabel("epoch")

plt.legend(["train","test"],loc="upper left")

plt.show()
```

```python

pred = model.predict_classes(X_test)
```

```python
from sklearn.metrics import  confusion_matrix, accuracy_score,roc_curve,auc


cm = confusion_matrix(y_test, pred)
acc = accuracy_score(y_test, pred)
```

```python

df_cm = pd.DataFrame(cm, index= ("low", "high"), columns= ("low", "high"))

sb.heatmap(df_cm, annot= True, fmt = 'g')
```

```python
def plot_roc(labels, predict_prob):
    false_positive_rate,true_positive_rate,thresholds=roc_curve(labels, predict_prob)
    roc_auc=auc(false_positive_rate, true_positive_rate)
    plt.title('ROC')
    plt.plot(false_positive_rate, true_positive_rate,'b',label='AUC = %0.4f'% roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0,1],[0,1],'r--')
    plt.ylabel('TPR')
    plt.xlabel('FPR')
plot_roc(y_test, pred)
```
